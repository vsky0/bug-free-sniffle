## Data Cleaning

1. Remove unrelevant columns
2. Convert `date` from string into datetime object
3. keep the high performing teams.

### Dropping columns

```python
df.columns
```

```python
cols_to_drop = ['mid','venue', 'batsman','bowler','striker','non-striker']

print(f"Shape of the dataframe before dropping columns: {df.shape}")

df.drop(labels=cols_to_drop, axis=1, inplace=True)
print(f"Shape of the dataframe after dropping columns: {df.shape}")
```

```python
df.columns
```

```python
df.sample(10)
```

```python
df['bat_team'].unique()
```

keeping the top teams.

```python
top_performing_teams = ['kkr','csk','rr','mi','kxip','rcb','dd','srh']
```

```python
top_teams = ['Kolkata Knight Riders', 'Chennai Super Kings', 'Rajasthan Royals',
                    'Mumbai Indians', 'Kings XI Punjab', 'Royal Challengers Bangalore',
                    'Delhi Daredevils', 'Sunrisers Hyderabad']
```

```python
print(f"Shape of dataframe before revmoving inconsistent teams: {df.shape}")
df = df[(df['bat_team'].isin(top_teams)) & (df['bowl_team'].isin(top_teams))]
print(f"Shape of dataframe after revmoving inconsistent teams: {df.shape}")
```

```python
df['bat_team'].unique()
```

remove the first 5 overs data

```python
df.head()
```

```python
df = df[df['overs']>=5.0]
print(f"Shape of dataframe after revmoving first 5 overs data: {df.shape}")
```

formatting date in string type to datetime object

```python
from datetime import datetime

df['date'] = df['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d'))
print(f"after converting the date to datetime object {type(df.iloc[0,0])}")
```

## Data Preprocessing

- handling categorical features
- splitting the dataset

```python
encoded_df = pd.get_dummies(data=df, columns=['bat_team','bowl_team'])
encoded_df.columns
```

```python
encoded_df.head()
```

```python
encoded_df = encoded_df[['date', 'bat_team_Chennai Super Kings', 'bat_team_Delhi Daredevils', 'bat_team_Kings XI Punjab',
              'bat_team_Kolkata Knight Riders', 'bat_team_Mumbai Indians', 'bat_team_Rajasthan Royals',
              'bat_team_Royal Challengers Bangalore', 'bat_team_Sunrisers Hyderabad',
              'bowl_team_Chennai Super Kings', 'bowl_team_Delhi Daredevils', 'bowl_team_Kings XI Punjab',
              'bowl_team_Kolkata Knight Riders', 'bowl_team_Mumbai Indians', 'bowl_team_Rajasthan Royals',
              'bowl_team_Royal Challengers Bangalore', 'bowl_team_Sunrisers Hyderabad',
              'overs', 'runs', 'wickets', 'runs_last_5', 'wickets_last_5', 'total']]
```

```python
xtrain = encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year <= 2016]
xtest = encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year >=2017]

ytrain = encoded_df[encoded_df['date'].dt.year <=2016]['total'].values
ytest = encoded_df[encoded_df['date'].dt.year>=2017]['total'].values

# remove date, its not useful now
xtrain.drop(labels='date',axis=1,inplace=True)
xtest.drop(labels='date',axis=1,inplace=True)

print(f"Training set:{xtrain.shape}")
print(f"Testing set: {xtest.shape}")
```

## Model Building

### Linear Regression

```python
from sklearn.linear_model import LinearRegression

linear_regression_model = LinearRegression()
linear_regression_model.fit(xtrain,ytrain)
```

```python
linear_regression_predictions = linear_regression_model.predict(xtest)
```

```python
from sklearn.metrics import mean_absolute_error as mae, mean_squared_error as mse

print("Linear Regression Model:")
print("MAE:",mae(ytest,linear_regression_predictions))
print("MSE:",mse(ytest,linear_regression_predictions))
print("RMSE",np.sqrt(mse(ytest,linear_regression_predictions)))
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score as acc # you don't need accuracy for the regression model stupid.

# dtr decision tree regressor 
dtr_model = DecisionTreeRegressor()
dtr_model.fit(xtrain,ytrain)
```

```python
dtr_predictions = dtr_model.predict(xtest)
```
